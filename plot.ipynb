Plot accuracy graph collected from various experiment into one graph

```python
import matplotlib.pyplot as plt

accuracy = {
    'baseline': {4: (212, 724, 150), 8: (316, 692, 219), 16: (311, 710, 222), 32: (319, 732, 227), 64: (305, 738, 218)},
    'dropout': {4: (215, 718, 158), 8: (320, 708, 218), 16: (298, 715, 207), 32: (296, 738, 210), 64: (296, 740, 211)},
    'layer_norm': {4: (123, 706, 84), 8: (253, 762, 191), 16: (304, 805, 240), 32: (347, 817, 273), 64: (344, 815, 268)},
    'residual': {4: (234, 747, 181), 8: (356, 737, 257), 16: (357, 750, 258), 32: (333, 775, 247), 64: (321, 781, 235)},
    'residual_stride': {4: (263, 703, 184), 8: (337, 715, 235), 16: (349, 745, 252), 32: (328, 766, 242), 64: (322, 778, 237)},
    'deep': {4: (303, 729, 217), 8: (319, 825, 259), 16: (336, 884, 286), 32: (320, 859, 264), 64: (285, 847, 233)},
    'deep_layer_norm': {4: (130, 661, 94), 8: (300, 712, 204), 16: (303, 838, 246), 32: (322, 848, 262), 64: (316, 843, 255)},
    'transformer': {4: (273, 682, 206), 8: (406, 718, 288), 16: (373, 716, 261), 32: (380, 725, 270), 64: (368, 726, 254)},
    'vae_deep': {4: (304, 712, 214), 8: (415, 746, 297), 16: (345, 822, 278), 32: (324, 845, 267), 64: (320, 833, 256)},
    'vae_transformer': {4: (299, 696, 227), 8: (408, 722, 292), 16: (416, 729, 296), 32: (379, 746, 268), 64: (340, 776, 249)},
}

keys = list(accuracy.keys())
x = list(accuracy['baseline'].keys())

def plot_one(idx, title):
    for k in keys:
        acc = [x[idx] / 1000 for x in accuracy[k].values()]
        plt.plot(x, acc, label=k)

    plt.xlabel('Embedding Size')
    plt.ylabel('Accuracy')
    plt.xscale('log')
    plt.xticks(x, x)
    plt.title(title)
    plt.legend()
    plt.show()

plot_one(0, 'Color Accuracy')
plot_one(1, 'Shape Accuracy')
plot_one(2, 'Color + Shape Accuracy')
```
